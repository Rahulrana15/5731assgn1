```python
# You code here (Please add comments in the code):
import nltk

from bs4 import BeautifulSoup as bs
import pandas as pd
import requests as rq
import matplotlib.pyplot as plt
import nltk
from wordcloud import WordCloud
from nltk.corpus import stopwords


#collecting sample text data from IMDb "godfather movie reviews"

#assigining the URL to a varilable 
url = 'https://www.imdb.com/title/tt13130308/reviews/?ref_=tt_ql_urv'

#Creating a data frame with a column to store text of reviews
d_f = pd.DataFrame(columns=['Review'])
req = rq.get(url).text
soup = bs(req,'html.parser')

#getting reviews
review_text = soup.find_all('div', attrs={'class' : 'text show-more__control'})

#creating a empty list
review_list = []
for i in range(len(review_text)):
    review_list.append(review_text[i].get_text())

#assigining list to the data frame    
d_f['Review'] = review_list

d_f['word_count'] = d_f['Review'].apply(lambda x: len(str(x).split(" ")))
d_f[['Review','word_count']].head()


```

```python
d_f['char_count'] = d_f['Review'].str.len() ## this also includes spaces
d_f[['Review','char_count']].head()
```

```python
#Stopwords

from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
stop = stopwords.words('english')

d_f['stopwords'] = d_f['Review'].apply(lambda x: len([x for x in x.split() if x in stop]))
d_f[['Review','stopwords']].head()
```

```python
d_f['hastags'] = d_f['Review'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
d_f[['Review','hastags']].head()
```

```python
d_f['numerics'] = d_f['Review'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
d_f[['Review','numerics']].head()
```

```python
d_f['upper'] = d_f['Review'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
d_f[['Review','upper']].head()
```
